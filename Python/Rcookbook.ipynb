```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
BOD = pd.read_csv("BOD.csv",index_col=[0])

fig, ax =plt.subplots(1,2,sharex=True,figsize=(15,5))
for i in (0,1):
  ax[i].plot(BOD['Time'],BOD['demand'])
  ax[i].set_xlabel("Time");ax[i].set_ylabel("demand")
  ax[i].grid(linestyle=':')

ax[1].set_ylim(0,max(BOD['demand']))
```

```python
plt.close()

worldpop = pd.read_csv('worldpop.csv',index_col=[0])

fig, ax =plt.subplots(1,2,sharex=True,figsize=(15,5))
for i in (0,1):
  ax[i].plot(worldpop['Year'],worldpop['Population'],marker='o')
  ax[i].set_xlabel("Year");ax[i].set_ylabel("Population")
  ax[i].grid(linestyle=':')

ax[1].set_yscale('log')
```

```python
plt.close()

toothgrowth = pd.read_csv('toothgrowth.csv',index_col=[0])
too_group = toothgrowth.pivot_table(values='len',index='supp',\
                                    columns='dose',aggfunc=np.mean)

fig, ax =plt.subplots(1,2,sharex=True,figsize=(15,5))

too_group.T.plot(ax=ax[0])
too_group.T.plot(ax=ax[1],color='black',style=['-',':'])
```

```python
plt.close()

climate = pd.read_csv('climate.csv',index_col=[0])
clim = climate[climate['Source'] == 'Berkeley'][['Year','Anomaly10y','Unc10y']]

plt.plot(clim['Year'],clim['Anomaly10y'])
plt.fill_between(clim['Year'], clim['Anomaly10y']-clim['Unc10y']\
                 , clim['Anomaly10y']+clim['Unc10y'],\
                 color='grey', alpha=.25)
```

```python
plt.close()

heightweight = pd.read_csv('heightweight.csv',index_col=[0])[['sex','ageYear','heightIn']]
hw = heightweight.groupby('sex')

fig,ax = plt.subplots(4,1,sharey=True,figsize=(5,15))
loc = [0,1,2,3]

for i in loc:
  for n,g in hw: 
    if n == "f":
      c = "red"
      s = "o"
    else:
      c = "blue"
      s = "^"
    if i == 0:
      ax[i].scatter(g.ageYear,g.heightIn,marker="o",color=c,label=n,s=20)
    elif i == 1:
      ax[i].scatter(g.ageYear,g.heightIn,marker=s,color='black',label=n,s=20)
    elif i == 2:
      ax[i].scatter(g.ageYear,g.heightIn,marker=s,color=c,label=n,s=20)
    else:
      ax[i].scatter(g.ageYear,g.heightIn,marker=s,edgecolors=c,label=n,s=20,facecolors="none")
  ax[i].set_xlabel("ageYear") ; ax[i].set_ylabel("heightIn")
  ax[i].legend(title="sex",bbox_to_anchor=(1.05, 1), loc='upper left')
  ax[i].grid(linestyle=":")
```
