```python
def get_category(age):
  cat = ''
  if age<20: cat='10대'
  elif age<30: cat='20대'
  elif age<40: cat='30대'
  elif age<50: cat='40대'
  elif age<60: cat='50대'
  elif age<70: cat='60대'
  else: cat='70세 이상'

  return cat
```

```python
df['Age_year'] = df['Age'].apply(lambda x : get_category(x))
df[['Age', 'Age_year']].head(5)
```
