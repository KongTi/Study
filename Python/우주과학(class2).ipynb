```python
import numpy as np
import matplotlib.pyplot as plt

L = [2,4,6,8]
theta = np.linspace(-1/2 * np.pi,1/2 * np.pi,1000, endpoint=True)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, polar=True)

def Graph(i):
  ax.plot(theta,i * (np.cos(theta))**2, label= 'L='+str(i)+'('+str(round(np.degrees(np.arccos(np.sqrt(1/i))),1))+')')

for i in L:
  Graph(i)

ax.plot(np.linspace(0, 2*np.pi, 100), np.ones(100),linewidth=3,color='black',label='Earth')
ax.set_xlim(-1/2 * np.pi,1/2 * np.pi)
ax.legend()
```
