```python
import numpy as np
from matplotlib import pyplot as plt

e = 1.6 * 10**-19
m = 9.1 * 10**-31
epslion = 8.85 * 10**-12
n = np.linspace(10**-6,10**-3,10**4)

def calculate():
  cof = 1/(2*np.pi)
  main = np.sqrt((n*e**2)/(m*epslion))
  return cof * main
 
fig, ax = plt.subplots(1,2,figsize=(15,6))
ax[0].plot(n*10**6,calculate() * 10**3)
ax[0].set_xlim(1,1000) ; ax[0].set_ylim(1,300)
ax[0].set_xlabel(r'$cm^{-3} $',fontsize=15) ; ax[0].set_ylabel(r'$kHz $',fontsize=15)
ax[0].set_title('Linear Scale',fontsize=20)

ax[1].plot(n*10**6,calculate() * 10**3)
ax[1].set_xlim(1,1000) ; ax[1].set_ylim(1,1000)
ax[1].set_xlabel(r'$cm^{-3} $',fontsize=15)
ax[1].set_xscale('log') ; ax[1].set_yscale('log')
ax[1].set_title('Log Scale',fontsize=20)
```
