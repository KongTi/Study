```python
import os

#path= os.path.dirname(os.path.abspath(__file__)) # 현재 파일 경로 얻기
path = "C:/Users/hjm01/Desktop/대학원/AI를 위한 기초 프로그래밍"
os.chdir(path) # 경로 변경

file_list = os.listdir() # 파일 리스트 얻기
file_txt = [file for file in file_list if file.endswith(".txt")][0] # .txt로 끝나는 텍스트 파일만 필터링
txt = open(file_txt,'r').readlines() # 파일 불러오기

# [문제 1]

for i in txt:
    print(i.strip('\n'))

# [문제 2]

print("\n파일이름 : {}".format(file_txt.split('.')[0]),'\n')
for i in txt:
    print(i.strip('\n'))

# [문제 3]

print("\n파일이름 : {}".format(file_txt.split('.')[0]))
for i in txt:
    ch_list = i.split()
    print('')
    print(i.strip('\n'))
    for j in ch_list:
        print(j)

# [문제 4]

print("\n파일이름 : {}".format(file_txt.split('.')[0]))
for i in txt:
    ch_list = i.split()
    print('')
    print(i.strip('\n'))
    for j in ch_list:
        print(j.strip('.'))
```
