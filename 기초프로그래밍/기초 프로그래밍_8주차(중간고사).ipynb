```python
# 문제 1

print("문자열을 입력하시오")
string = str(input())
print("3번째~5번째 문자열")
print(string[2:5])

# 문제 2

print("문자열을 입력하시오")
string = str(input())
print("시작 문자 위치")
start = int(input())
print("종료 문자 위치")
end = int(input())
print("{}번째~{}번째 문자열".format(start,end))
print(string[start-1:end])

# 문제 3

print("문자열을 입력하시오")
string = str(input())
string = string.replace(" ","")
element = sorted(list(''.join(set(string))))
for i in element:
    num = string.count(i)
    print("문자 '{}'의 개수 : {}개".format(i,num))

# 문제 4

k = 1 ; str_list = []
for i in range(6):
    print("{}번째 문자열을 입력하시오".format(k))
    string = str(input())
    str_list.append(string)
    print(str_list)

    k += 1
front = list([j[0] for j in str_list])
print('acronym : {}'.format("".join(front)))

# 문제 5

peptide = ['His','Phe','Tyr','Leu','Pro','Met']
cost = [4, 7, 3, 5, 8, 1]
print("peptide = {}".format(peptide))
print("cost = {}".format(cost))
print("개수를 입력하시오")
num = int(input())

sum_cost = 0
for i in range(num):
    print("{}번째 문자열을 입력하시오".format(i+1))
    string = str(input())
    try:
        index = peptide.index(string)
    except ValueError:
        print("없는 문자열입니다 (cost = 0)")
        continue
    sum_cost += cost[index]

print('cost : {}'.format(sum_cost))
```
