```python
# [문제 1]

import urllib.request
page = urllib.request.urlopen("https://www.ncbi.nlm.nih.gov/genbank/samplerecord/")
text = page.read().decode("utf8")

start_of_gene = text.find("ORIGIN") + 10
end_of_gene = text.find("FIELD") - 43

gene_txt = str(text[start_of_gene:end_of_gene])

print("aga의 갯수 : "+str(gene_txt.count("aga")))

# [문제 2]

print("\n<aga -> tag>")
print(gene_txt.replace("aga","tag"))
```
