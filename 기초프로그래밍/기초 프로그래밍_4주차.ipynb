```python
# [문제1]

# 1
list = []
print("1:",list)

# 2
add_element = ['apple','mango','banana','melon','tomato']
for i in add_element:
    list.append(i)
print("2:",list)

# 3
del list[-1]
print("3:",list)

# 4
list.insert(0,'kiwi')
print("4:",list)

# 5
print("5:",list[:])

# 6
print("6:",list[::2])

# 7
print("7:",list[2])

# [문제 2]

# 1
fruit = ['orange','banana','avocado','melon','Grapes']
print("1:",fruit)

# 2
color = ['black','blue','green','white','red']
print("2:",color)

# 3
for count,i in enumerate(zip(fruit,color)):
    if count == 0:
        print("3:","fruit : {}, color : {}".format(i[0],i[1]))
    else:
        print("fruit : {}, color : {}".format(i[0],i[1]))

# 4-1
try:
    idx1 = int(input("fruit에서 출력할 원소의 인덱스 번호를 입력하세요 : "))
    print("4-1:",fruit[idx1])
except IndexError:
    print("Error : 0~{}까지의 숫자중에 선택해주세요".format(fruit.index(fruit[-1])))

# 4-2
try:
    idx2 = int(input("color에서 출력할 원소의 인덱스 번호를 입력하세요 : "))
    print("4-2:",color[idx2])
except IndexError:
    print("Error : 0~{}까지의 숫자중에 선택해주세요".format(color.index(color[-1])))

# 5
idx3 = int(input("5보다 작은 정수 n을 입력하세요 : "))
if idx3 < 5:
    print("5:",fruit[:idx3+1]); print(color[:idx3+1])
else:
    print("5보다 작은 정수를 입력해주세요")
```
